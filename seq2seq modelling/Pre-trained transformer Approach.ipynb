```python
from datasets import load_dataset

dataset=load_dataset("agentlans/en-fr")

print(dataset)
```

```python
train_data=dataset["train"].shuffle(seed=42).select(range(20000))
test_data=dataset["test"].shuffle(seed=42).select(range(7000))

split=test_data.train_test_split(test_size=0.5,seed=42)
val_data=split["train"]
test_data=split["test"]
```

```python
from transformers import T5Tokenizer, T5ForConditionalGeneration

model_name = "t5-small"
tokenizer = T5Tokenizer.from_pretrained(model_name)
model = T5ForConditionalGeneration.from_pretrained(model_name)
```

```python
print(train_data)
```

```python
MAX_LEN = 90

def preprocess(example):
    source = "translate English to French: " + example["english"]
    target = example["french"]

    inputs = tokenizer(source, max_length=MAX_LEN, padding="max_length", truncation=True)
    targets = tokenizer(target, max_length=MAX_LEN, padding="max_length", truncation=True)

    inputs["labels"] = targets["input_ids"]
    return inputs

train_tok = train_data.map(preprocess, remove_columns=train_data.column_names)
val_tok = val_data.map(preprocess, remove_columns=val_data.column_names)
test_tok = test_data.map(preprocess, remove_columns=test_data.column_names)
```

```python
from transformers import DataCollatorForSeq2Seq
from torch.utils.data import DataLoader

collator = DataCollatorForSeq2Seq(tokenizer, model=model)

train_loader = DataLoader(train_tok, batch_size=8, shuffle=True, collate_fn=collator)
val_loader = DataLoader(val_tok, batch_size=8, shuffle=False, collate_fn=collator)
test_loader = DataLoader(test_tok, batch_size=8, shuffle=False, collate_fn=collator)
```

```python
import torch
from tqdm import tqdm

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)

optimizer = torch.optim.AdamW(model.parameters(), lr=5e-4)

for epoch in range(3):
    model.train()
    total_loss = 0
    loop = tqdm(train_loader, desc=f"Epoch {epoch+1}")

    for batch in loop:
        input_ids = batch["input_ids"].to(device)
        attention_mask = batch["attention_mask"].to(device)
        labels = batch["labels"].to(device)

        outputs = model(input_ids=input_ids, attention_mask=attention_mask, labels=labels)
        loss = outputs.loss

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        total_loss += loss.item()
        loop.set_postfix(loss=loss.item())

    print(f"Epoch {epoch+1} Avg Loss: {total_loss / len(train_loader):.4f}")
```

```python
import evaluate
metric = evaluate.load("sacrebleu")

model.eval()
preds = []
refs = []

for batch in tqdm(val_loader, desc="Evaluating"):
    input_ids = batch["input_ids"].to(device)
    attention_mask = batch["attention_mask"].to(device)

    with torch.no_grad():
        generated = model.generate(input_ids=input_ids, attention_mask=attention_mask, max_length=128)

    decoded_preds = tokenizer.batch_decode(generated, skip_special_tokens=True)
    decoded_refs = tokenizer.batch_decode(batch["labels"], skip_special_tokens=True)

    preds.extend(decoded_preds)
    refs.extend([[ref] for ref in decoded_refs])  # nested list for sacreBLEU

score = metric.compute(predictions=preds, references=refs)
print("Validation BLEU Score:", score["score"])
```

```python
from tqdm import tqdm

metric = evaluate.load("sacrebleu")

model.eval()
test_preds = []
test_refs = []

for batch in tqdm(test_loader, desc="Evaluating on Test"):
    input_ids = batch["input_ids"].to(device)
    attention_mask = batch["attention_mask"].to(device)

    with torch.no_grad():
        generated = model.generate(input_ids=input_ids, attention_mask=attention_mask, max_length=128)

    decoded_preds = tokenizer.batch_decode(generated, skip_special_tokens=True)
    decoded_refs = tokenizer.batch_decode(batch["labels"], skip_special_tokens=True)

    test_preds.extend(decoded_preds)
    test_refs.extend([[ref] for ref in decoded_refs]) 

# Compute BLEU
test_score = metric.compute(predictions=test_preds, references=test_refs)
print("Test BLEU Score:", test_score["score"])
```
